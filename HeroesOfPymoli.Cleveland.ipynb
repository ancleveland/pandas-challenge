```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file)
```

## Player Count

* Display the total number of players


```python
# Display total number of players. Need to determine unique players in order to count.

total_players = len(purchase_df["SN"].unique())
player_count = pd.DataFrame({"Total Players":[total_players]})
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Run basic calculations to obtain number of unique items, average price, total puchases and total revenue.
unique_items = len(purchase_df["Item ID"].unique())
average_price = purchase_df["Price"].mean()
num_of_purchases = purchase_df["SN"].count()
total_revenue = purchase_df["Price"].sum()

# Create a summary data frame to hold the results
summary_data_table = pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price], "Number of Purchases":[num_of_purchases], "Total Revenue":[total_revenue]})
# Optional: give the displayed data cleaner formatting - insert $ sign
summary_data_table["Average Price"] = summary_data_table["Average Price"].map("${:,.2f}".format)
summary_data_table["Total Revenue"] = summary_data_table["Total Revenue"].map("${:,.2f}".format)
                   
# Display the summary data frame
sum_of_purchases = summary_data_table[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
sum_of_purchases
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Percentage and Count of Male Players
# Percentage and Count of Female Players
# Percentage and Count of Other / Non-Disclosed
# Use groupby feature to group into gender
gender_group = purchase_df.groupby("Gender")
unique_gender = gender_group["SN"].nunique()
percentage_of_players = unique_gender/total_players

# Create a summary data frame to hold the results
gender_summary = pd.DataFrame({"Total Count": unique_gender,"Percentage of Players": percentage_of_players})
# Optional: give the displayed data cleaner formatting - insert $ sign
gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:,.2%}".format)
# Display the summary data frame
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
gender_group = purchase_df.groupby("Gender")
purchase_count = gender_group["SN"].count()
avg_purchase_price = gender_group["Price"].mean()
tot_purchase_value = gender_group["Price"].sum()
avg_purchase_total = (gender_group["Price"].sum()/unique_gender)

# Create a summary data frame to hold the results
purchase_summary = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":avg_purchase_price,
                           "Total Purchase Value":tot_purchase_value,"Avg Purchase Total per Person":avg_purchase_total})
# Optional: give the displayed data cleaner formatting
purchase_summary["Average Purchase Price"] = purchase_summary["Average Purchase Price"].map("${:,.2f}".format)
purchase_summary["Total Purchase Value"] = purchase_summary["Total Purchase Value"].map("${:,.2f}".format)
purchase_summary["Avg Purchase Total per Person"] = purchase_summary["Avg Purchase Total per Person"].map("${:,.2f}".format)
# Display the summary data frame
purchase_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
bins = [0,9,14,19,24,29,34,39,50]

# Create bin labels
age_bin = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_df["Age Ranges"] = pd.cut(purchase_df["Age"],bins, labels=age_bin)
purchase_df

# Calculate the numbers and percentages by age group - "Total Count" and "Percentage of Players"
age_group = purchase_df.groupby("Age Ranges")
unique_members = age_group["SN"].nunique()
percent_of_players = (unique_members/total_players)
# Create a summary data frame to hold the results
age_summary = pd.DataFrame({"Total Count":unique_members, "Percentage of Players":percent_of_players})

# Optional: round the percentage column to two decimal points
age_summary["Percentage of Players"] = age_summary["Percentage of Players"].map("{:,.2%}".format)

# Display Age Demographics Table
age_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Run basic calculations to obtain the results in the table below.
top_spenders = purchase_df.groupby(["SN"])
top_count = top_spenders["Item ID"].count()
top_tot_purch = top_spenders["Price"].sum()
top_avg_purch = top_spenders["Price"].mean()

# Create a summary data frame to hold the results
top_spend_summary = pd.DataFrame({"Purchase Count":top_count, "Average Purchase Price":top_avg_purch, "Total Purchase Value":top_tot_purch})

# Sort the total purchase value column in descending order
# try this format sample:final_df = df.sort_values(by=['2'], ascending=False)

top_spend_summary = top_spend_summary.sort_values("Total Purchase Value", ascending=False)

# Optional: give the displayed data cleaner formatting
top_spend_summary ["Average Purchase Price"] = top_spend_summary ["Average Purchase Price"].map("${:,.2f}".format)
top_spend_summary ["Total Purchase Value"] = top_spend_summary ["Total Purchase Value"].map("${:,.2f}".format)

# Display a preview of the summary data frame
top_spend_summary.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain puchase count, item price, and total purchase price
popular_items = purchase_df.groupby(["Item ID", "Item Name"])
item_count = popular_items["SN"].count()
item_price = popular_items["Price"].mean()
item_total = popular_items["Price"].sum()

# Create a summary data frame to hold the results
items_summary = pd.DataFrame({"Purchase Count":item_count, "Item Price":item_price, "Total Purchase Value":item_total})

# Sort the purchase count column in descending order
items_summary = items_summary.sort_values("Purchase Count", ascending=False)

# Optional: give the displayed data cleaner formatting
items_summary ["Item Price"] = items_summary ["Item Price"].map("${:,.2f}".format)
items_summary ["Total Purchase Value"] = items_summary ["Total Purchase Value"].map("${:,.2f}".format)
# Display a preview of the summary data frame
items_summary.head(5)
```
